# Exploring Health Data for Barbados

To explore the data I will be using the folium library. You need to install this on the computer running python. Then import the library. 

We create a base map centered on the island with an appropiate zoom level and tiles overlay. 

```python
import folium

map = folium.Map(location=[13.1742, -59.5581], zoom_start=11, tiles="Stamen Terrain")
```

The next step is to create a feature group. Here I am using a geojson file from the website

https://data.humdata.org/dataset/barbados-healthsites/resource/0d52c842-e8ef-42dd-b9f3-4ba78b5b3cd7
    
Then run the map with the add_child function to include the feature group. This displays the map with the geojson overlay.   

```python
fg_health = folium.FeatureGroup(name="Health Data")

fg_health.add_child(folium.GeoJson(data=open('barbados_healthsites.geojson', 'r', encoding='utf-8-sig').read(),))

map.add_child(fg_health)
```

Folio allows you to save the map to an html file. You can now use this in your website. Very cool.

```python
map.save("barbadosmap.html")
```
